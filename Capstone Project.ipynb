# **Comovement**

```
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv("CapstoneProject - CapstoneProject (1).csv")

plt.plot(data['Year'],data['Gold_Mumbai'],label="Gold_Mumbai")
plt.plot(data['Year'],data['Gold_London'],label="Gold_London")
plt.plot(data['Year'],data['BSE_Index'],label="BSE_Index")
plt.plot(data['Year'],data['Oil_Price'],label="Oil_Price")

plt.legend()
plt.xlabel("Year")
plt.show()
```
