```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
#it is the coronavirus-data of february month from 01-02-2020 to 10-02-2020. it has four columns date,new_cases,new_deaths_total_cases

#its a toy data
data = {'date': ['2020-02-01','2020-02-02','2020-02-03','2020-02-04','2020-02-05','2020-02-06','2020-02-07','2020-02-08',
                 '2020-02-09','2020-02-10'], 
        'new_cases':[21,26,28,32,39,37,32,34,26,29],
        'new_deaths': [6,4,5,6,6,7,7,8,8,9], 
        'new_recover': [9, 4, 17, 26, 22, 8, 26, 18, 7, 44],}
df = pd.DataFrame(data, columns = ['date','new_cases', 'new_deaths', 'new_recover']) 
df['date']=pd.to_datetime(df['date'])
df.info()
df=df.set_index('date')
df.info()
plt.plot('new_cases',color='red',marker='o',label='new_cases',data=df)
plt.plot('new_deaths',color='green',marker='v',label='new_deaths',data=df)
plt.plot('new_recover',color='yellow',marker='*',label='new_recover',data=df)
plt.xticks(rotation=70) 
plt.xlabel("date")
plt.ylabel("No. of patients")
plt.legend()
plt.show()
```
